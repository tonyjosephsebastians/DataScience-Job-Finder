```python
import spacy

# Load the English language model in spaCy
nlp = spacy.load('en_core_web_sm')

# Define a function to extract technical skills from job descriptions
def extract_technical_skills(description):
    # Parse the job description with spaCy
    doc = nlp(description)

    # Extract all the named entities in the job description
    entities = [entity.text for entity in doc.ents]

    # Filter out entities that are not technical skills
    technical_skills = []
    for entity in entities:
        if entity.lower() in TECHNICAL_SKILLS:
            technical_skills.append(entity)

    # Return the list of technical skills
    return technical_skills

# Define a list of technical skills to search for
TECHNICAL_SKILLS = ['python', 'java', 'c++', 'sql', 'javascript', 'html', 'css', 'git', 'aws', 'azure','spark','tableau']

# Extract the top technical skills from the job descriptions
top_technical_skills = {}
for description in df['job_description']:
    skills = extract_technical_skills(description)
    for skill in skills:
        if skill in top_technical_skills:
            top_technical_skills[skill] += 1
        else:
            top_technical_skills[skill] = 1

# Sort the skills by frequency in descending order
top_technical_skills = dict(sorted(top_technical_skills.items(), key=lambda item: item[1], reverse=True))

# Print the top 10 skills
for skill, frequency in list(top_technical_skills.items())[:10]:
    print(skill, frequency)

```

```python
import urllib.request
import json
import folium
import pandas as pd

# Get the count of jobs in each city
city_counts = df['company_location'].value_counts()

# Get the top 10 cities (excluding 'Canada')
top_cities = city_counts[city_counts.index.str.contains('Canada') == False].nlargest(10)

# Create a DataFrame with city, country, and job count information
city_info = pd.DataFrame({'city': top_cities.index, 'count': top_cities.values})
city_info['country'] = city_info['city'].apply(lambda x: x.split(',')[-1].strip())

# Use geopy to get the latitude and longitude of each city
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="my-app")

city_info['location'] = city_info['city'].apply(geolocator.geocode)

city_info['latitude'] = city_info['location'].apply(lambda loc: loc.latitude if loc else None)
city_info['longitude'] = city_info['location'].apply(lambda loc: loc.longitude if loc else None)

# Drop any rows with missing latitude or longitude values
city_info = city_info.dropna(subset=['latitude', 'longitude'])

# Add top 5 job titles to each city
for index, row in city_info.iterrows():
    jobs = df[df['company_location'] == row['city']]['job_title'].value_counts().nlargest(5)
    city_info.at[index, 'top_jobs'] = ', '.join(jobs.index)

url = 'https://raw.githubusercontent.com/nvkelso/natural-earth-vector/master/geojson/ne_110m_admin_0_countries.geojson'

with urllib.request.urlopen(url) as url:
    countries_geojson = json.loads(url.read().decode())

country_info = []
for feature in countries_geojson['features']:
    properties = feature.get('properties', {})
    country_name = properties.get('name')
    if not country_name:
        continue
    for index, row in city_info.iterrows():
        if row['country'] == country_name:
            properties['count'] = row['count']
            properties['top_jobs'] = row['top_jobs']
            country_info.append(feature)
            break

# Create a new GeoJSON file with the job counts for each country
jobs_geojson = {'type': 'FeatureCollection', 'features': country_info}

# Create a map centered on North America
m = folium.Map(location=[45, -100], zoom_start=3)

# Add the choropleth layer
folium.Choropleth(
    geo_data=jobs_geojson,
    name='choropleth',
    data=city_info,
    columns=['country', 'count'],
    key_on='feature.properties.name',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Job Counts'
).add_to(m)

# Add markers for the top 10 cities
for index, row in city_info.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=row['city'] + ': ' + str(row['count']) + ' jobs<br>' + row['top_jobs']
    ).add_to(m)

# Add a layer control to the map
folium.LayerControl().add_to(m)

# Display the map
m

```

```python


import urllib.request
import json
import pandas as pd
import plotly.express as px

# Get the count of jobs in each city
city_counts = df['company_location'].value_counts()

# Get the top 10 cities (excluding 'Canada')
top_cities = city_counts[city_counts.index.str.contains('Canada') == False].nlargest(10)

# Create a DataFrame with city, country, and job count information
city_info = pd.DataFrame({'city': top_cities.index, 'count': top_cities.values})
city_info['country'] = city_info['city'].apply(lambda x: x.split(',')[-1].strip())

# Use geopy to get the latitude and longitude of each city
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="my-app", timeout=10)

city_info['location'] = city_info['city'].apply(geolocator.geocode)

city_info['latitude'] = city_info['location'].apply(lambda loc: loc.latitude if loc else None)
city_info['longitude'] = city_info['location'].apply(lambda loc: loc.longitude if loc else None)

# Drop any rows with missing latitude or longitude values
city_info = city_info.dropna(subset=['latitude', 'longitude'])

# Add top 5 job titles to each city
for index, row in city_info.iterrows():
    jobs = df[df['company_location'] == row['city']]['job_title'].value_counts().nlargest(5)
    city_info.at[index, 'top_jobs'] = ', '.join(jobs.index)

url = 'https://raw.githubusercontent.com/nvkelso/natural-earth-vector/master/geojson/ne_110m_admin_0_countries.geojson'

with urllib.request.urlopen(url) as url:
    countries_geojson = json.loads(url.read().decode())

country_info = []
for feature in countries_geojson['features']:
    properties = feature.get('properties', {})
    country_name = properties.get('name')
    if not country_name:
        continue
    for index, row in city_info.iterrows():
        if row['country'] == country_name:
            properties['count'] = row['count']
            properties['top_jobs'] = row['top_jobs']
            country_info.append(feature)
            break

# Create a new GeoJSON file with the job counts for each country
jobs_geojson = {'type': 'FeatureCollection', 'features': country_info}

# Create a Plotly choropleth map
fig = px.choropleth_mapbox(city_info, geojson=jobs_geojson, color='count', locations='country',
                           featureidkey='properties.name',
                           center={'lat': 45, 'lon': -100},
                           mapbox_style="carto-positron", zoom=3,
                           opacity=0.7, labels={'count': 'Job Counts'})

# Add markers for each city
fig.add_scattermapbox(
    lat=city_info['latitude'],
    lon=city_info['longitude'],
    text=city_info['top_jobs'],
    mode='markers',
    marker=dict(size=8, color='red', opacity=0.8)
)

# Display the map
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Filter the DataFrame to include only the top 10 cities
top_cities = df['company_location'].value_counts().nlargest(10).index
filtered_df = df[df['company_location'].isin(top_cities)]

# Group the filtered DataFrame by job title and company location, and count the number of occurrences for each group
grouped_df = filtered_df.groupby(['job_title', 'company_location']).size().reset_index(name='count')

# Sum the counts of job postings for each job title and select the top 5 job titles based on the highest sums
top_job_titles = grouped_df.groupby('job_title')['count'].sum().nlargest(5).index

# Create a new DataFrame that contains the counts of job postings for each top job title and top company location
top_jobs_df = pd.DataFrame(columns=['job_title', 'company_location', 'count'])
for job_title in top_job_titles:
    job_df = grouped_df[grouped_df['job_title'] == job_title]
    top_locations = job_df.sort_values(by='count', ascending=False)['company_location'][:10]
    top_jobs_df = top_jobs_df.append(job_df[job_df['company_location'].isin(top_locations)])

# Create subplots for each job title
fig, axs = plt.subplots(nrows=len(top_job_titles), sharex=True, figsize=(8, 16))

# Iterate over the job titles and create a bar chart for each one
for i, job_title in enumerate(top_job_titles):
    job_df = top_jobs_df[top_jobs_df['job_title'] == job_title]
    city_counts = job_df.set_index('company_location')['count'].reindex(top_cities, fill_value=0)
    axs[i].barh(city_counts.index, city_counts.values)
    axs[i].set_title(job_title)
    axs[i].set_yticklabels(city_counts.index)

# Set shared labels for the subplots
fig.text(0.5, 0.04, 'Number of job postings', ha='center')


# Set the overall title for the chart
plt.suptitle('Top 5 job titles in top 10 cities')

plt.show()
```
